# DATAFRAMES PARA MODELO ENTIDAD-RELACIÓN

### TABLA "CATEGORY"

```python
category = pd.DataFrame(columns=["Category"])
category["Category"] = total_data["Category"].unique()
category["Category_Id"] = category.index + 1
order = ['Category_Id', 'Category']
Category = category[order]
Category
```

### TABLA "CITY"

```python
## Diccionario con los datos de los estados
datos = {
    'State': ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware', 'Florida', 'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan', 'Minnesota', 'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire', 'New Jersey', 'New Mexico', 'New York', 'North Carolina', 'North Dakota', 'Ohio', 'Oklahoma', 'Oregon', 'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia', 'Washington', 'West Virginia', 'Wisconsin', 'Wyoming'],
    'State_Init': ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY']
}

# Crear el DataFrame
df_states = pd.DataFrame(datos)
df_states.head(2)
```

```python
city = total_data[["City", "State"]]
city = city.drop_duplicates()
city = city.drop(city[city["City"] == ''].index)
city['State'] = city['State'].apply(lambda x: df_states.loc[df_states['State_Init'] == x, 'State'].values[0] if x in df_states['State_Init'].values else x)
city = city.sort_values(by='City', ascending=True)
city
```

```python
city = city.reset_index(drop = True)
city["City_Id"] = city.index + 1
order = ['City_Id', 'City', "State"]
city = city[order]
city['State'] = city['State'].str.replace('_', ' ')
city
```

```python
states = df_states
states["State_Id"] = states.index + 1
order = ['State_Id', 'State', "State_Init"]
states = states[order]
states.head(2)
```

```python
fc = pd.merge(city, states, on="State", how="left")
fc['State_Id'] = fc['State_Id'].fillna(0).astype(int)
fc
```

```python
## Tabla City
City = fc[["City_Id", "City", "State_Id"]]
City.head(10)
```

### TABLA "STATE"

```python
## Tabla State
State = fc[["State_Id", "State", "State_Init"]]
State = State.drop_duplicates()
State = State.dropna(subset=['State_Init'])
State = State.sort_values(by='State_Id', ascending=True)
State = State.reset_index(drop = True)
State
```

### TABLA "USER"

```python
user = total_data[["User_Id", "User_Name"]]
user = user.drop_duplicates()
User = user.reset_index(drop = True)
User
```

#### TABLA "BUSINESS"

```python
business = total_data[["Business_Id", "Business_Name", "Address", "City", "State", "Category", "Latitude", "Longitude"]]
business = business.drop_duplicates()
business['State'] = business['State'].apply(lambda x: df_states.loc[df_states['State_Init'] == x, 'State'].values[0] if x in df_states['State_Init'].values else x)
business.head(3)
```

```python
bs = pd.merge(business, Category, on="Category", how="left")
bs = pd.merge(bs, State, on="State")
bs = bs.drop_duplicates()
bs = pd.merge(bs, City, on=["City", "State_Id"])
bs
```

```python
## Tabla Facts_Business
Business = bs[["Business_Id", "Business_Name", "Address", "Category_Id", "City_Id", "Latitude", "Longitude"]]
Business
```

#### TABLA "REVIEW"

```python
review = total_data[["Business_Id", "Date", "User_Id", "Category","Ranking", "Stars", "Text"]]
review = review.reset_index(drop = True)
review["Review_Id"] = review.index + 1
review = review.drop_duplicates()
review
```

```python
rv = pd.merge(review, Category[["Category_Id", "Category"]], on="Category", how="inner")
rv
```

```python
Facts_Review = rv[["Review_Id", "Business_Id", "Date", "User_Id", "Category_Id", "Ranking",  "Stars", "Text"]]
Facts_Review.head(10)
```

#### TABLA "CHECKIN"

```python
checkin = total_data[["Business_Id", "Checkin_Count", "Date"]]
checkin = checkin.drop_duplicates()
checkin = checkin.reset_index(drop = True)
checkin["Checkin_Id"] = checkin.index + 1
checkin
```

```python
Facts_Checkin = checkin[["Checkin_Id", "Business_Id", "Checkin_Count"]]
Facts_Checkin
```
